An example of Python/Numpy broadcasting



```python
import numpy as np

#From the above Matrix we have
#             apple beef egg   potatos
A = np.array([[56., 0.,   4.4, 68.],
              [1.2, 104., 52., 8.],
              [1.8, 135., 99., 0.9]])

print(A)
```

```python
# Total calories, sums down columns
cal = A.sum(axis=0)
print(cal)
```

```python
# The reshape command is not really necessary here but can use it to
percentage = 100* A/cal.reshape(1,4)
print(percentage)

#reshape example
a = np.arange(6)
print(f'a = {a}')

b= a.reshape(3,2)
print(f'b = {b}')
```
